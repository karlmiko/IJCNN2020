# GTzan Dataset

http://marsyasweb.appspot.com/download/data_sets/

- The dataset consists of 1000 audio tracks each 30 seconds long.

- It contains 10 genres, each represented by 100 tracks.

- The tracks are all 22050Hz Mono 16-bit audio files in .wav format.

```python
import numpy as np
import os
import soundfile as sf
import math

import matplotlib.pyplot as plt
import pylab as py
from scipy import fftpack
from scipy import signal
from pylab import *
import scipy.io.wavfile
```

```python
# Path to source files
path2='../genres'
```

## List all wav files

```python
# List of all wav files
file_list = []
i = 0 
for path, subdirs, files in os.walk(path2):
    print(shape(files))
    for name in files:
        if (name[-2:] != 'au'):
            continue
        file_list.append(os.path.join(path, name))
        i += 1
print("Files processed: "+str(i) )
```

## Get the label (genre) from directory name

```python
# Get the label (genre) from directory name and create a list of genres

genre_list  = list()

track_id = 0

for file in file_list:
    
    genre = str(file.split('/')[2])
    print(genre)
    genre_list.append(genre)
    track_id += 1
```

```python
shape(genre_list), shape(file_list)
```

## Create 3 data folds and directories (fold1, fold2, fold3) and write the wav files into them

```python
# Split the Dataset into 3 Folds
from sklearn.model_selection import StratifiedKFold

gkf = StratifiedKFold(n_splits=3, random_state=7, shuffle=True)
gkf.split(file_list, genre_list)
print(gkf)
fold_num = 1 

for train_index, test_index in gkf.split(file_list, genre_list):
    
    print("\nTRAIN:", train_index, "\n\nTEST:", test_index)
    print("\nTRAIN:", shape(train_index), "TEST:", shape(test_index))
    
    for ii in test_index:
        data , samplerate = sf.read(file_list[ii])
        genre = file_list[ii].split('/')[2]
        track = file_list[ii].split('/')[3]
        directory = "gtzan/fold"+str(fold_num)+"/"+genre
        #Verify if it exists, otherwise create it
        if not os.path.exists(directory):
            os.makedirs(directory)
            
        print("gtzan/fold"+str(fold_num)+"/"+genre+"/"+track)
        sf.write((directory+"/"+track), data, samplerate)
    fold_num +=1
    
    
```
