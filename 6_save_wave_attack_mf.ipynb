```python
#Imports

import librosa
import math
import librosa.display
import IPython.display
import numpy as np
import os
import soundfile as sf
```

```python

nb_classes = 10
sr = 22050
frame_size = 110250
```

```python
# Put all filenames into a list

train_fold = [1, 2]
test_fold = [3]
```

```python

test_filename_list  = []
base_path = "gtzan_3f_Win_110250_75/"

j  = 0 
ii = 0
for ii in test_fold:

    path = base_path+"fold"+str(ii)

    for path, subdirs, files in os.walk( path ):
        for name in files:
            test_filename_list.append(os.path.join(path, name))
            j += 1
    print("Test  files processed: "+"fold"+str(ii)+" "+str(j) )
```

```python
f = test_filename_list[24]
g = f.split('/')
print(str(g[1] + "/" + g[2] + "/" + g[3]))
```

```python
# Load 5 second Original Mel Spec

X_valid         = np.load( "folds_mf/2_GTzan_Xs_test_fold"+str(test_fold[0])+"_110250_75_frozen.npy")
X_attacked_bim  = np.load( "folds_mf/2_GTzan_Xs_BIM_fold"+str(test_fold[0])+"_110250_75_frozen.npy")
X_attacked_fgsm = np.load( "folds_mf/2_GTzan_Xs_FGSM_fold"+str(test_fold[0])+"_110250_75_frozen.npy")
X_sample = X_valid[9:10]  #Arbitrary value
```

```python

X_sample = np.squeeze(X_sample)
print(X_sample.shape)

len_X = X_valid.shape[0]
print(len_X)
```

```python
#Reconstructing original mel and attacked mel specs

head="gtzan_reconstruct/"
i = 1110;

for filename in test_filename_list[1110:]:
    
    original_power_mel = librosa.db_to_power(np.squeeze(X_valid[i,:]))    
    audio_rec_original = librosa.feature.inverse.mel_to_audio(original_power_mel, sr=sr, n_fft=512, power=1.0, hop_length=256, fmin=0.0, fmax=sr/2)
    
    attacked_power_mel_bim = librosa.db_to_power(np.squeeze(X_attacked_bim[i,:]))
    audio_rec_attacked_bim = librosa.feature.inverse.mel_to_audio(
        attacked_power_mel_bim, sr=sr, n_fft=512, power=1.0, hop_length=256, fmin=0.0, fmax=sr/2)
    
    attacked_power_mel_fgsm = librosa.db_to_power(np.squeeze(X_attacked_fgsm[i,:]))
    audio_rec_attacked_fgsm = librosa.feature.inverse.mel_to_audio(
        attacked_power_mel_fgsm, sr=sr, n_fft=512, power=1.0, hop_length=256, fmin=0.0, fmax=sr/2)
    
    
    # Save original (reconstructed from Mel) and attacked audio files
    
    f = test_filename_list[i]
    g = f.split('/')
    path = str(g[1] + "/" + g[2] + "/")
    
    directory_mf          = head + "mf/"      + path
    directory_attack_bim  = head + "mf_BIM/"  + path
    directory_attack_fgsm = head + "mf_FGSM/" + path
    
    if not os.path.exists(directory_mf):
            os.makedirs(directory_mf)
    if not os.path.exists(directory_attack_bim):
            os.makedirs(directory_attack_bim)
    if not os.path.exists(directory_attack_fgsm):
            os.makedirs(directory_attack_fgsm)
    
    sf.write(directory_mf          + g[3], audio_rec_original,      sr)
    sf.write(directory_attack_bim  + g[3], audio_rec_attacked_bim,  sr)
    sf.write(directory_attack_fgsm + g[3], audio_rec_attacked_fgsm, sr)
    
    if (i%10 == 0):
            print("Done with:", i, "out of", len_X)
    i += 1
    
```
